```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler

```

```python
# Load the dataset
df = pd.read_csv('D:/Sukrutha_Datasets/housing.csv')
df.head()
```

```python
# Select features for clustering
X = df[['median_income', 'housing_median_age']]

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)
y_pred = kmeans.predict(X_scaled)

# Plot the clusters
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_pred)
plt.xlabel('Median Income')
plt.ylabel('Median House Age')
plt.title('KMeans Clustering')
plt.show()
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
# Select features for clustering
X = df[['median_income', 'housing_median_age']]

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
dbscan = DBSCAN(eps=12.5,min_samples = 4).fit(X_scaled)
labels = dbscan.labels_
```

```python
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=labels, cmap="plasma")
plt.xlabel('Median Income')
plt.ylabel('Housing Median Age')
plt.show()
```

# Hierarchial Clustering

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Select features for clustering
X = df[['median_income', 'housing_median_age']]
# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#perform hierarchial clustering
Z = linkage(X_scaled, method ='ward')
```

```python
plt.figure(figsize = (10,5))
plt.xlabel('Sample index')
plt.ylabel('distance')
dendrogram(Z)
plt.show()
```
